```python
import pandas as pd
import numpy as np
```

```python
# Đọc Dataframe từ file csv, với các cột như sau
field_names = ['State', 'Sex', 'Year', 'Name', 'Count']
babynames_df = pd.read_csv('Data/STATE.CA.TXT',header=None,names=field_names)
```

```python
# Show 10 dòng ngẫu nhiên
babynames_df.sample(10)
```

```python
# Hiển thị DF chỉ gồm female 
female_df = babynames_df[babynames_df['Sex'] == 'F']
female_df
```

```python
# Hiển thị DF gồm những người male, sinh trước 2000
male_2000_df = babynames_df[(babynames_df['Sex'] == 'M') & (babynames_df['Year']<2000) ]
male_2000_df
```

```python
# Hiển thị 3 babynames với count > 250
filtered_df = babynames_df[babynames_df['Count']>250]
filtered_df.head(3)
```

```python
# Hiển thị các babynames có tên là Bella, Alex, Narges, hoặc Lisa
selected_names = ['Bella', 'Alex', 'Narges', 'Lisa']
filtered_df = babynames_df[babynames_df['Name'].isin(selected_names)]
filtered_df
```

```python
# Hiển thị với những Name bắt đầu bằng chữ N
filtered_df = babynames_df[babynames_df['Name'].str.startswith('N')]
filtered_df
```

```python
# Thêm cột name_lengths với giá trị là số lượng ký tự của cột Name
babynames_df['name_lengths'] = babynames_df['Name'].apply(len)
babynames_df
```

```python
# Hãy cho biết số lượng baby có tên 'Alice' trung bình hàng năm là bao nhiêu
alice_df = babynames_df[babynames_df['Name']=='Alice']

average_count_per_year = alice_df.groupby('Year')['Count'].mean()

print("Số lượng baby có tên 'Alice' trung bình hàng năm:")
print(average_count_per_year)
```

```python
# Hiển thị tất cả các tên trong DF, mỗi tên chỉ hiển thị 1 lần
unique_names = babynames_df['Name'].drop_duplicates()
unique_names
```

```python
# Tìm babynames phổ biến nhất trong năm 2020
babynames_2020_df = babynames_df[babynames_df['Year'] == 2020]
most_popular_names_2020 = babynames_2020_df[babynames_2020_df['Count'] == babynames_2020_df['Count'].max()]

print("Babynames phổ biến nhất trong năm 2020:")
print(most_popular_names_2020)
```
