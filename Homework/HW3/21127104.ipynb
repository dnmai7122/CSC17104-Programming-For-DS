## Exercises 1

Use `re.search` to find whether a string contains a phone number. The pattern that you write should detect a phone number in the following strings.  
```
"Call me at 382-384-3840."  
"my number is (510) 849-3519. Call me!"
```  
And not find a match in the following strings. 
```
"my number is 510-849-35192"  
"here’s my number: 510-849.3519"
``` 
Consider making your own tests as well  

````python
# # YOUR CODE HERE
import re

def find_phone_number(text):
    # Revised pattern to match phone numbers in the specified format
    # Removing word boundary check around the area code in parentheses
    pattern = r'(?:\(\d{3}\)\s?\d{3}-\d{4}|\b\d{3}-\d{3}-\d{4}\b)'
    match = re.search(pattern, text)
    if match:
        return match.group()
    else:
        return None

# Test strings
test_strings = [
    "Call me at 382-384-3840.",
    "my number is (510) 849-3519. Call me!",
    "my number is 510-849-35192",
    "here’s my number: 510-849.3519"
]

for text in test_strings:
    phone_number = find_phone_number(text)
    if phone_number:
        print(f"Found a phone number in '{text}': {phone_number}")
    else:
        print(f"No phone number found in '{text}'")
````

## Exercise 2

Use `re.sub` to alter the string below so that the dates have a common format that uses a dash for the day, month, and year separator.  
```
03/12/2018, 03.13.18, 03/14/2018, 03:15:2018
```

````python
# YOUR CODE HERE
# Test strings
test_strings = "03/12/2018, 03.13.18, 03/14/2018, 03:15:2018"
# Define a regular expression pattern to match different date formats
pattern = r'(\d{2})[/.:](\d{2})[/.:](\d{2,4})'

# Use re.sub to replace the date formats with the common format (with dashes)
formatted_text = re.sub(pattern, r'\1-\2-\3', test_strings)

print(formatted_text)
````

## Exercise 4

In this exercise, you you working with ```email_test.txt``` file (attached), using Regular Expression. 
1. Extract name and email address of sender  
Output should be something like this

<u>**Noted**</u>: If the email don't have name, you can skip this it.
```
Name:
"MR. JAMES NGOLA." 
"Mr. Ben Suleman" 
"PRINCE OBONG ELEME" 
"PRINCE OBONG ELEME" 
"Maryam Abacha" 
```
```
Email address:
"james_ngola2002@maktoob.com"
"bensul2004nng@spinfinder.com"
"obong_715@epatra.com"
"obong_715@epatra.com"
"m_abacha03@www.com"
```
2. Find SUBJECT of All email with Date from 01/06/2003 to 13/05/2004 </br>
<u>**Noted**</u>:
- If the email don't have Date, you can skip this it.
- SUBJECT may appear before or after Date 

````python
import re
fh = open(r"Data/email_test.txt", "r").read()

# YOUR CODE HERE
# 1. Extract name and email address of sender
# 2. Find SUBJECT of All email from 01/06/2003 to 13/05/2004

from datetime import datetime


# 1. Extract name and email address of sender
sender_info_pattern = r'From: "([^"]+)" <([^>]+)>'
sender_info_matches = re.findall(sender_info_pattern, fh)

sender_names = [name for name, _ in sender_info_matches]
sender_emails = [email for _, email in sender_info_matches]

print("Name:")
for name in sender_names:
    print(name)

print("\nEmail address:")
for email in sender_emails:
    print(email)

# 2. Find SUBJECT of All emails from 01/06/2003 to 13/05/2004
email_sections = re.split(r'\n(?=From:)', fh)

# Convert string dates to datetime objects for comparison
start_date = datetime.strptime("01/06/2003", "%d/%m/%Y")
end_date = datetime.strptime("13/05/2004", "%d/%m/%Y")

for email_section in email_sections[1:]:
    date_match = re.search(r'Date: .+ (\d{2} \w{3} \d{4})', email_section)
    subject_match = re.search(r'Subject: (.+)', email_section)
    
    if date_match and subject_match:
        email_date_str = date_match.group(1)
        email_subject = subject_match.group(1)

        # Here we handle multiple possible date formats
        for date_format in ("%d %b %Y", "%a, %d %b %Y %H:%M:%S", "%d/%m/%Y"):  # Add more formats as needed
            try:
                email_date = datetime.strptime(email_date_str, date_format)
                if start_date <= email_date <= end_date:
                    print(f"Date: {email_date_str}\nSubject: {email_subject}\n")
                break  # If we successfully parse the date, we can break the loop
            except ValueError:
                continue  # Try the next date format
````
