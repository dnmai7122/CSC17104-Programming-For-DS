## Đưa ra câu hỏi cần trả lời

Sau khi đã hiểu hơn về dữ liệu, ta sẽ quay lại bước đưa ra câu hỏi cần trả lời: tinh chỉnh câu hỏi, hoặc có thể đưa ra thêm các câu hỏi mới.

Câu hỏi “tình trạng hiểu về Python của các bạn trong lớp?” có thể được cụ thể hóa ra thành 2 câu hỏi:

2. Câu quiz nào có nhiều bạn làm sai nhất, câu nào có nhiều bạn làm sai nhì, ...?

## Tiền xử lý + phân tích dữ liệu để trả lời cho câu hỏi 1

*Câu hỏi 1: Cột điểm tổng “Grade/10.00” có phân bố các giá trị như thế nào?*

Ở đây không cần tiền xử lý mà sẽ đi luôn vào phân tích dữ liệu.

Đầu tiên, ta sẽ xem thử median của cột "Grade/10.00".

```python
grades = cols['Grade/10.00']
sorted_grades = sorted(grades)
sorted_grades
p= 75 #p=25,75 là bách trung vị
i= int(p/100 * (len(sorted_grades) -1))
print(sorted_grades[i])
```

Để cảm thụ rõ hơn về phân bố các giá trị của cột "Grade/10.00", ta sẽ chia miền giá trị của cột này thành các khoảng (bin) và xem thử mỗi khoảng có bao nhiêu sinh viên.

Giả sử ta sẽ chia đoạn [0, 10] thành 5 bin như sau: [0, 2], (2, 4], [4, 6), [6, 8), [8, 10]. Ta có thể lưu kết quả vào một dictionary, trong đó: key là một tuple cho biết 2 đầu của một bin (thay tuple bằng list thì có được không?), value là số lượng sinh viên rơi vào bin này.



```python
bin_ranges = [(0, 2), (2, 4), (4, 6), (6, 8), (8, 10)]
bin_counts = {bin_range: 0 for bin_range in bin_ranges}

for grade in grades:
    for bin_range in bin_ranges:
        if bin_range[0] <= grade < bin_range[1] or (bin_range[1] == 10 and grade == 10):
            bin_counts[bin_range] += 1
            break

print(bin_counts)
```

## Tiền xử lý + phân tích dữ liệu để trả lời cho câu hỏi 2

*Câu hỏi 2: Câu quiz nào có nhiều bạn làm sai nhất, câu nào có nhiều bạn làm sai nhì, ...?*

Ta có thể lưu kết quả vào một dictionary, trong đó: key là tên của cột ứng với câu quiz, value là điểm trung bình của cột. Dictionary này được sort tăng dần theo value.



```python
import csv

data = []
with open('Data/PythonQuiz2023.csv', 'r', encoding='utf-8-sig') as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        data.append(row)

average_scores = {}

for column in data[0].keys():
    if column.startswith("Q. "):
        total_score = 0
        count = 0
        for row in data:
            value = row[column]
            if value != "-" and value != "":
                try:
                    total_score += float(value)
                    count += 1
                except ValueError:
                    pass
        if count > 0:
            average_score = total_score / count
            average_scores[column] = average_score

sorted_average_scores = dict(sorted(average_scores.items(), key=lambda item: item[1], reverse=True))

best_question, best_score = next(iter(sorted_average_scores.items()))
worst_question, worst_score = list(sorted_average_scores.items())[-1]  # Lấy phần tử cuối cùng

print(f"Câu hỏi có điểm trung bình cao nhất là {best_question} với điểm {best_score}")
print(f"Câu hỏi có điểm trung bình thấp nhất là {worst_question} với điểm {worst_score}")
```

```python
# In giá trị trung bình của từng câu hỏi ra màn hình
for question, average_score in average_scores.items():
    print(f"Câu hỏi {question}: Điểm trung bình = {average_score}")
```
